```python
# import tools 
import pandas as pd 
import seaborn as sns 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
fname = 'BayArea_df.xlsx'
df1 = pd.read_excel(fname, sheet_name = 0)
```

```python
df1.head()
```

```python
plt.figure(figsize=(16, 6))
sns.kdeplot(data=df1,x = 'Unemployment Pctl', hue= 'California County')
```

```python
plt.figure(figsize=(16, 6))
sns.kdeplot(data=df1,x = 'Low Birth Weight Pctl', hue= 'California County')
```

```python
plt.figure(figsize=(16, 6))
sns.kdeplot(data=df1,x = 'Ozone Pctl', hue= 'California County')
```

```python
sns.jointplot(x = "Ozone Pctl", y='Low Birth Weight Pctl', hue= 'California County', data=df1, kind = 'kde')
```

```python
plt.figure(figsize=(16, 6))
ax = sns.boxplot(x="California County", y="Ozone Pctl", data=df1)
```
